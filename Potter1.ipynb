```python
import pandas as pd
from datascience import *
import numpy as np
import matplotlib.pyplot as plots
%matplotlib inline
```

```python
from urllib.request import urlopen 
import re
def read_url(url): 
    return re.sub('\\s+', ' ', urlopen(url).read().decode())
```

```python
url = "https://raw.githubusercontent.com/amephraim/nlp/refs/heads/master/texts/J.%20K.%20Rowling%20-%20Harry%20Potter%201%20-%20Sorcerer's%20Stone.txt"
# Read the data
data = pd.read_fwf(url, header=None)
```

```python
huck_finn_url = 'https://inferentialthinking.com/data/huck_finn.txt'
huck_finn_text = read_url(huck_finn_url)
huck_finn_chapters = huck_finn_text.split('CHAPTER ')[44:]
```

```python
harry_potter_url = 'https://raw.githubusercontent.com/amephraim/nlp/refs/heads/master/texts/J.%20K.%20Rowling%20-%20Harry%20Potter%201%20-%20Sorcerer%27s%20Stone.txt'
harry_potter_text = read_url(harry_potter_url)
harry_potter_chapters = harry_potter_text.split('CHAPTER ')[1:]
```

```python

# Count how many times the names Jim, Tom, and Huck appear in each chapter

counts = Table().with_columns([
        'Jim', np.char.count(huck_finn_chapters, 'Jim'),
        'Tom', np.char.count(huck_finn_chapters, 'Tom'),
        'Huck', np.char.count(huck_finn_chapters, 'Huck')
    ])

# Plot the cumulative counts:
# how many times in Chapter 1, how many times in Chapters 1 and 2, and so on.

cum_counts = counts.cumsum().with_column('Chapter', np.arange(1, 44, 1))
cum_counts.plot(column_for_xticks=3)
plots.title('Cumulative Number of Times Name Appears');
```

```python
# Lets do the same for Harry Potter
counts_potter = Table().with_columns([
        'Harry', np.char.count(harry_potter_chapters, 'Harry'),
        'Ron', np.char.count(harry_potter_chapters, 'Ron'),
        'Hermione', np.char.count(harry_potter_chapters, 'Hermione')
    ])
counts_potter
```

```python
cum_counts = counts_potter.cumsum().with_column('Chapter', np.arange(1, 18, 1))
cum_counts.plot(column_for_xticks=3)
plots.title('Cumulative Number of Times Name Appears');
```
